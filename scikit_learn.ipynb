```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

content = load_breast_cancer()
X_train, X_test, y_train, y_test=train_test_split(content.data, content.target, random_state=42)

model = DecisionTreeClassifier(criterion='entropy')
model
```

```python
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
pred=model.predict(X_test)
pred_2d=pred.reshape(len(pred),1)
pred_2d
```

```python
import numpy as np
y_test_2d = y_test.reshape(len(pred), 1)
df1 = pd.DataFrame(pred_2d)
df2 = pd.DataFrame(y_test_2d)

df_concat = pd.concat([df1,df2],axis=1)

df_concat.columns=['pred', 'real']
df_concat[df_concat['pred'] == df_concat['real']]
```
